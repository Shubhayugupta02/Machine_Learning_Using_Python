```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/AIML_PRACTICE/CarPrice_Assignment.csv")
df
```

```python
df=df.drop(['car_ID','CarName'],axis=1)
df
```

```python
x=df.iloc[:,:23].values
y=df.iloc[:,23].values
x
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
x[:,1]=en.fit_transform(x[:,1])
x[:,2]=en.fit_transform(x[:,2])
x[:,3]=en.fit_transform(x[:,3])
x[:,4]=en.fit_transform(x[:,4])
x[:,5]=en.fit_transform(x[:,5])
x[:,6]=en.fit_transform(x[:,6])
x[:,12]=en.fit_transform(x[:,12])
x[:,13]=en.fit_transform(x[:,13])
x[:,15]=en.fit_transform(x[:,15])
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
rmse=mse**0.5
print("Root Mean Square error=",rmse)
mae=mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error=",mae)
```
